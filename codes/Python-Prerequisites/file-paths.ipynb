```python
example_path = "example1.txt"
if os.path.exists(example_path):
    print(f"The path '{example_path}' exists")
else:
    print(f"The path '{example_path}' does not exists")
```

```python
# cheking if a path is file or directory
example_path = "example1.txt"
if os.path.isfile(example_path):
    print("It is a file")
elif os.path.isdir(example_path):
    print("It is a directory")
else:
    print("The path is neither a directory or a file")
```

```python
### Getting the absolute path
relative_path = "example.txt"
abs_path = os.path.abspath(relative_path)
print(abs_path)
```
