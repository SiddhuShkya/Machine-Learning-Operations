#### How to create custom package/module

1. Create a new folder (package)
2. Create "__init__.py" inside the folder.
3. Create a python file inside the folder which becomes your custom package.
4. Define your function inside that python file/package.

```python
from package.maths import addition

addition(1,2,3,4,5)
```
