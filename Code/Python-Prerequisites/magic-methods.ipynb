```python
## Basic Methods
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age  
        

person = Person("Siddhu", 24)
print(person)
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
    def __str__(self):
        return f"Person(name={self.name}, age={self.age})"
    
    def __repr__(self):
        return f"Person(name={self.name}, age={self.age})"


person = Person("Siddhu", 24)
print(person)
print(repr(person))
```
