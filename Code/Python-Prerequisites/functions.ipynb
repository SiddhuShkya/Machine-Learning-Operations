#### Variable length arguments

```python
# Positional and Keyword Arguments

def print_nums(*args, **kwargs):
    print("args : ", args)
    for arg in args:
        print(f"Positional argument: {arg}")
    print("kwargs : ", kwargs)    
    for key, value in kwargs.items():
        print(f"Keyword argument: {key} = {value}")
```

```python
print_nums(1, 2, 3, name="Alice", age=30)
```

#### Lambda Functions

```python
addition = lambda a,b: a + b  # noqa: E731
type(addition)
```

```python
addition(5, 7)  # returns 12
```

```python
even = lambda num: num % 2 == 0  # noqa: E731
even(4)  # returns True
```

```python
numbers = [1,2,3,4,5]
square = lambda x:x**2  # noqa: E731
squared_numbers = list(map(square, numbers))
print(squared_numbers)  # prints [1, 4, 27, 256, 3125]
```

#### Map function

```python
def square(num):
    return num * num
```

```python
nums = [1, 2, 3, 4, 5]
map(square, nums)  # returns a map object
res = list(map(square, nums))
print(res)  # prints [1, 4, 9, 16,
```

```python
# lamda function with map function
numbers = [1,2,3,4,5]
squared_numbers = list(map(lambda x: x**2, numbers))  # noqa: E731
print(squared_numbers)  # prints [1, 4, 9, 16, 25]
```

```python
# map multiple iterables
list1 = [1, 2, 3]
list2 = [4, 5, 6]
summed_list = list(map(lambda x, y: x + y, list1, list2))  # noqa: E731
print(summed_list)  # prints [5, 7, 9]  
```

```python
# map to convert a list of strings to uppercase
strings = ["hello", "world", "python"]
uppercase_strings = list(map(lambda s: s.upper(), strings))  # noqa: E731
print(uppercase_strings)  # prints ['HELLO', 'WORLD', 'PYTHON']
```

```python
# map to do type conversion
str_numbers = ["1", "2", "3", "4", "5"]
int_numbers = list(map(int, str_numbers))   
print(int_numbers)  # prints [1, 2, 3, 4, 5]
```

```python
# map to a list of dictionaries

def get_names(persons):
    return persons['name']

persons = [
    {'name': "siddhu", 'age': 23},
    {'name': "roman",'age': 25}
]

list(map(get_names, persons))
```
