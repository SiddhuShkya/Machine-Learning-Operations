#### Object Oriented Programming

```python
# A class is a blue print for creating objects. Attributes, methods

class Dog:
    
    def __init__(self, name, age): # constructor
        self.name = name           # Instance variables
        self.age = age
    
    def bark(self):                # Instance method
        print(f"{self.name} says woof")
```

```python
d1 = Dog(name="Luna", age=2)
d2 = Dog(name='Hero', age=8)
```

```python
print(d1.name)
print(d1.age)
```

```python
print(d2.name)
print(d2.age)
```

```python
d1.bark()
```

```python
## Modeling a Bank Account

class BankAccount:
    
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance
        
    def deposit(self, amount):
        self.balance += amount
        print(f"{amount} has been deposited. New Balance : {self.balance}")
        
    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient funds")
            return
        self.balance -= amount
        print(f"{amount} has been withdrawn. New Balance : {self.balance}")
        
    def get_balance(self):
        return self.balance
```

```python
acc = BankAccount(owner="Siddhartha", balance=1000)
```

```python
acc.deposit(500)
```

```python
acc.withdraw(800)
```

```python
acc.get_balance()
```

#### Inheritance

```python
## Single Inheritance
# Parent Class
class Car:
    
    def __init__(self, windows, doors, enginetype):
        self.windows = windows
        self.doors = doors
        self.enginetype = enginetype
        
    
    def drive(self):
        print(f"The person wil drive the {self.enginetype} car")
        

# Child class
class Tesla(Car):
    
    def __init__(self, windows, doors, enginetype, is_selfdriving):
        super().__init__(windows, doors, enginetype)    ## calling init form the parent class
        self.is_selfdriving = is_selfdriving
        
    def self_driving(self):
        if self.is_selfdriving:
            print("Tesla supports self driving")
            return 
        print("Tesla does not supports self driving")


        
```

```python
c1 = Car(windows=4, doors=2, enginetype='petrol')
c1.drive()
```

```python
c2 = Tesla(windows=2, doors=2, enginetype='electric', is_selfdriving=False)
c2.self_driving()
```

```python
c2.drive()
```

```python
## Multiple Inheritance
## when a class inherits from more than one base class.

class Animal: # Base Class 1
    
    def __init__(self, name):
        self.name = name
        
    def speak(self):
        print("Subclass must implement this method")
        

class Pet: # Base Class 2
    
    def __init__(self, owner):
        self.owner = owner
        
        
class Dog(Animal, Pet): # Derived class
    
    def __init__(self, name, owner):
        Animal.__init__(self, name)
        Pet.__init__(self, owner)
        
        
    def speak(self):
        return f"{self.name} says woof!!"
```

```python
dog = Dog(name='Luna', owner="Siddhartha")
print(dog.speak())
print("Owner : ", dog.owner)
```

#### Polymorhism

```python
## Method overiding

# Base class
class Animal:
    
    def speak(self):
        return "SOund of the animal"
    
# Derived class 1
class Dog(Animal):
    
    def speak(self):
        return "Woof!"
    
# Derived class 2
class Cat(Animal):
    
    def speak(self):
        return "Meow!"
```

```python
dog = Dog()
cat = Cat()
```

```python
print(dog.speak())
print(cat.speak())
```

```python
# Fucntion that demonstrates polymorphism
def animal_speak(animal):
    print(animal.speak())
    
animal_speak(dog)
```

```python
# Polymorhism with functions and methods

## Base class
class Shape:
    
    def area(self):
        return "The area of the figure"

## Derived class 1
class Rectangle(Shape):
    
    def __init__(self, width, height):
        self.width = width
        self.height = height
        
        
    def area(self):
        return self.width * self.height
    

## Derived class 2
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius
    

## Function that demonstrates polymorhism

def print_area(shape):
    print(f"The area is {shape.area()}")
```

```python
rectangle = Rectangle(width=5, height=10)
circle = Circle(radius=7)

print_area(rectangle)
print_area(circle)
```

```python
## Polymorphism With abstract base classes (Interface)
from abc import ABC, abstractmethod

# Define Abstract Class
class Vehicle(ABC):
    
    @abstractmethod
    def start_engine(self):
        pass
    
    
# Derived class 1
class Car(Vehicle):
    
    def start_engine(self):
        return "Car engine started"
    
# Derived class 2
class Motorcycle(Vehicle):
    
    def start_engine(self):
        return "Motorcycle engine started"
    
def start_vehicle(vehicle):
    print(vehicle.start_engine())
```

```python
# create objs of car and motorcycle

car = Car()
motorcycle = Motorcycle()

start_vehicle(car)
```

### Encapsulation and Abstraction

```python
## Encapsulation with getter and setter
## Public, protected, private variables or access modifiers

class Person:
    
    def __init__(self, name, age, gender):
        self.__name = name # private variables
        self._age = age     # protected variables
        self.gender = gender # public variable
    
    # getter method for name
    def get_name(self):
        return self.__name
    
    #setter method for name
    def set_name(self, name):
        self.__name = name 
        

    
class Employee(Person):
    
    def __init__(self, name, age, gender):
        super().__init__(name, age, gender) 
        
        
        
person = Person('Siddhu', 23, 'male')
```

```python
person.gender
```

```python
person.set_name("Roman")
print(person.get_name())
```

```python
employee = Employee('Siddhartha', 25, 'male')
employee._age
```

```python
## Abstraction : Hiding complex implementation & showing only necessary features of an object

from abc import ABC, abstractmethod


# Abstract Class
class Vehicle(ABC):
    @abstractmethod
    def start(self):
        pass

    @abstractmethod
    def stop(self):
        pass


# Concrete Class 1
class Car(Vehicle):
    def start(self):
        print("Car engine started")

    def stop(self):
        print("Car engine stopped")


# Concrete Class 2
class Bike(Vehicle):
    def start(self):
        print("Bike engine started")

    def stop(self):
        print("Bike engine stopped")
```

```python
car = Car()
car.start()
car.stop()
```

```python
bike = Bike()
bike.start()
bike.stop()
```
